```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
%matplotlib inline
```

```python
k = np.array([5,10,25,50])
ovb = np.log(np.array([0.43, 0.532, 0.6, 0.73]))
gibbs = np.log(np.array([0.55, 0.61, 0.68, 0.75]))
```

```python
plt.clf()
plt.plot(k, ovb, 'r', label="OVB")
plt.plot(k, gibbs, 'b',label="Gibbs")
plt.legend()
plt.xlabel('Number of Topics')
plt.ylabel('Log-likelihood')
plt.show()
```

```python
k = np.array([5,10,25,50])
gibbs = np.array([61, 71, 122.5, 183.1])
ovb = np.array([21, 24, 32, 43])
```

```python
plt.clf()
plt.plot(k, ovb, 'r', label="OVB")
plt.plot(k, gibbs, 'b',label="Gibbs")
plt.legend()
plt.xlabel('Number of Topics')
plt.ylabel('Time for Posterior Estimation (minutes)')
plt.show()
```

```python
time = np.array([1,2,3])
fifteen = np.log(np.array([0.392, 0.3821, 0.3603]))
thirty = np.log(np.array([0.429, 0.483, 0.46]))
fifty = np.log(np.array([0.433, 0.46, 0.48]))
```

```python
plt.clf()
plt.plot(time, fifteen, linestyle='--', marker='o', color='r', label="k=15")
plt.plot(time, thirty, linestyle='--', marker='o', color='b', label="k=30")
plt.plot(time, fifty, linestyle='--', marker='o', color='g',label="k=50")
plt.legend(loc=2)
plt.xlabel('Time Slice')
plt.ylabel('Log-Likelihood')
plt.show()
```
